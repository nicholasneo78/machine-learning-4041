```python
from os import listdir
from os.path import isfile, join, splitext
import os
import pandas as pd
```

```python
FILEPATH = './Datasets/test'
```

```python
onlyfiles = [f for f in listdir(FILEPATH) if isfile(join(FILEPATH,f))]
```

```python
id_list = [splitext(k)[0] for k in onlyfiles]
```

```python
id_list
```

```python
# save the id into a csv
id_df = pd.DataFrame(id_list, columns=['id'])
id_df.to_csv('./Datasets/test_set_id.csv', index=False)
```
